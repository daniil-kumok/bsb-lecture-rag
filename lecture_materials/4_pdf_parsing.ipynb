# 4 — PDF Parsing

We need to get text out of PDF files before we can do anything with it.

We'll use **PyMuPDF** (imported as `fitz`) to extract text and see what comes out.

```python
import fitz  # this is PyMuPDF — pip package is called 'pymupdf'
```

## Open a PDF and check the basics

```python
doc = fitz.open("../docs/Travel.pdf")

print(f"Number of pages: {len(doc)}")
```

## Extract text from the first page

```python
page = doc[0]
text = page.get_text()

print(text)
```

## Extract all pages

```python
full_text = ""
for page in doc:
    full_text += page.get_text()

print(f"Total characters: {len(full_text)}")
print("---")
print(full_text)
```

## What's wrong with this output?

Look closely at the raw text above. You'll notice:

- **Random line breaks** in the middle of sentences — PDFs store text by position on the page, not by logical flow
- **Headers/footers** repeated on every page
- **No clear section boundaries** — where does one topic end and another begin?
- **Tables come out garbled** — columns get mashed together into one line

This raw text is not great input for an LLM. If you dump the whole thing into a prompt, the model will struggle to find the relevant part.

That's why we need **chunking** — splitting this mess into smaller, meaningful pieces.
