# 1 — API Calls

Before we do anything fancy, let's make sure we can talk to the Mistral API.

This notebook:
- Loads our API key from the `.env` file
- Creates a Mistral client
- Lists available models to confirm the connection works

```python
from dotenv import load_dotenv
from mistralai import Mistral
import os

load_dotenv()
```

## Create the client

The API key is stored in `.env` so we don't hardcode secrets in our code.

```python
client = Mistral(api_key=os.getenv("mistral_api_key"))
```

## List available models

A simple API call to verify everything is connected. If this works, we're good to go.

```python
resp = client.models.list()
print(resp)
```

That's a lot of output. Let's just print the model IDs:

```python
for model in resp.data:
    print(model.id)
```
